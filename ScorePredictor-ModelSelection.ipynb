## Dropping the irrelevant columns

```python
df.drop(columns=['mid','date','venue', 'batsman', 'bowler', 'striker', 'non-striker'], inplace=True)
```

## Keeping only the consistent teams

```python
current_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians','Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
df= df[(df['bat_team'].isin(current_teams)) &( df['bowl_team'].isin(current_teams))]
```

## Using One Hot Encoding for bat_team and bowl_team

```python
encoded_df= pd.get_dummies(df, columns=['bat_team', 'bowl_team'])
```

```python
encoded_df.head()
```

```python
#Splitting into Target and Independent Variables
x= encoded_df.drop(columns=['total'])
y= encoded_df['total']
```

```python
#Saving the Encoded Independent Variables(x) into a csv file
new_df= x.to_csv('encoded_df.csv', index=False)
```

```python
x.head()
```

## Splitting into train and test data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2)
```

```python
print(x_train.shape, x_test.shape)
```

## Scaling the Data

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled= scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#since no hyperparameter is being specified for linearRegression, it is not being included in the models_list below
model=LinearRegression()
model.fit(x_train_scaled, y_train)
y_pred=model.predict(x_test_scaled)
print(model, 'MAE: ', metrics.mean_absolute_error(y_test , y_pred), 'RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
scores=[]
scores.append([model,' ',metrics.mean_absolute_error(y_test , y_pred) , np.sqrt(metrics.mean_squared_error(y_test, y_pred))])
```

```python
#Defining various hyperparameters for each model 
models_list={
    'decison_tree':{
        'model':DecisionTreeRegressor(),
        'params':{
        'criterion':['squared_error','absolute_error'],
        'max_features':['sqrt', 'log2']
        }
    },
    'random_forest':{
        'model':RandomForestRegressor(),
        'params':{
            'n_estimators':[64,100,128],
            'max_features':['sqrt', 'log2']
        }
    }

  }
```

## Selecting the Best hyperparameters of each model and then choosing a model for our predictions

```python
from sklearn.model_selection import GridSearchCV
best_models = []

# Iterate through models_list and perform GridSearchCV directly
for model_name, config in models_list.items():
    model = config['model']
    params = config['params']
    
    grid_search = GridSearchCV(model, param_grid=params, cv=3, scoring='neg_mean_squared_error')
    grid_search.fit(x_train_scaled, y_train)
    
    best_models.append({
        'model_name': model_name,
        'best_model': grid_search.best_estimator_,
        'best_params': grid_search.best_params_,
        'best_score': grid_search.best_score_
    })

    
```

```python
for best_model_info in best_models:
    best_model = best_model_info['best_model']
    best_params = best_model_info['best_params']
    
    best_model.set_params(**best_params)
    best_model.fit(x_train_scaled, y_train)

    y_pred = best_model.predict(x_test_scaled)
    scores.append([best_model,best_params,metrics.mean_absolute_error(y_test , y_pred) , np.sqrt(metrics.mean_squared_error(y_test, y_pred))])
```

```python
#scores is a list which contains model name, mae and rmse.
#converting list to a dataframe
scores_df=pd.DataFrame(scores)
scores_df.rename(columns={0:'Model', 1:'HyperParameters',2:'MAE',3: 'RMSE'}, inplace=True)
scores_df
```

```python
#choosing the best model from above results
#Random Forest has the least error out of all 3 models

random_forest_model= RandomForestRegressor(n_estimators=128, max_features='log2')
random_forest_model.fit(x_train_scaled, y_train)
y_pred= random_forest_model.predict(x_test_scaled)
```
